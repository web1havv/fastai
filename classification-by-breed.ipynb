```python
import fastbook
fastbook.setup_book()

from fastbook import *

from fastai.vision.all import *
```

```python
path = untar_data(URLs.PETS)/'images'



# To see the image
img_files=get_image_files(path)
len(img_files)

import matplotlib.pyplot as plt
img = PILImage.create(img_files[3])
plt.imshow(img)
pat = r'^(.*)_\d+.jpg'
```

```python
## Pehle Path lena hai phir usse image nikalni hai phir PIL image create kar wani hai
```

```python
##Yaha par image from name re we we are focusing on file name
dls = ImageDataLoaders.from_name_re(
    path, get_image_files(path), pat, valid_pct=0.2, seed=42,
    item_tfms=Resize(244), batch_tfms=aug_transforms(size=224))
##batch_transfms isliye lagaya taaki pictures ke data achhe se aajaye ghuma phira ke bhi

dls.show_batch()

```

```python
learn = vision_learner(dls, resnet34, metrics=error_rate)
learn.fine_tune(1,3.630780702224001e-05)
```

```python
learn.lr_find()
```

```python
uploader = SimpleNamespace(data = ["/kaggle/input/sssss/Saint-Bernard-MP.jpg"])
     

img = PILImage.create(uploader.data[0])
breed = learn.predict(img)

img.show()
breed[0]
```
